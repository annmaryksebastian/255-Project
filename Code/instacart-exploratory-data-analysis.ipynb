##### Hypothesis Testing

There are significant differences between reorder proportions. Since we have assumed that the number of product orders is at least 50, the differences in reorder proportions between different products is statistically significant whenever the difference in reorder proportions is at least $0.20$ and the reorder proportions are between $0.2$ and $0.8$.

Define the pooled proportion, $\hat{p}$, the standard error, $SE$, and $z$-statistic by

\begin{equation*}
\hat{p} = \frac{x_1 + x_2}{n_1 + n_2}
\end{equation*}

\begin{equation*}
SE = \sqrt{\hat{p} (1-\hat{p}) \left(n_1^{-1} + n_2^{-1}\right)}
\end{equation*}

\begin{equation*}
z = \frac{\hat{p}_1 - \hat{p}_2}{SE}
\end{equation*}

where $\hat{p}_1$, $\hat{p}_2$ are reorder proportions; $n_1$, $n_2$ are total orders; and $x_1$, $x_2$ are reorders for two products. At a level of significance $\alpha = 0.05$, we need $|z| > 2$, i.e.
\begin{equation*}
2 < |z| = \left| \frac{\hat{p}_1 - \hat{p}_2}{SE} \right|
\end{equation*}

which, using the above formula for standard error, is
\begin{equation*}
2 \sqrt{\hat{p} (1-\hat{p}) \left(n_1^{-1} + n_2^{-1}\right)} < \left| \hat{p}_1 - \hat{p}_2 \right|.
\end{equation*}

We can minimize the left hand side to obtain the best lower bound for the difference in proportions $\left| \hat{p}_1 - \hat{p}_2 \right|$ by choosing $\hat{p} = 0.5$ and minimizing $n_1$ and $n_2$. Smaller sample sizes and smaller differences between sample statistics yield a smaller (in magnitude) $z$-statistic; for such samples we are less likely to adopt $H_1$ (stated below) than for larger samples or samples with larger differences in reorder proportions. When $n_1 = n_2 = 50$ and $\hat{p} = 0.5$,
\begin{equation*}
0.2 < \left| \hat{p}_1 - \hat{p}_2 \right|.
\end{equation*}

In order to construct hypothesis tests, compare the reorder proportions of a few pairs of products as a "spot check". We have assumed that the number of product orders is at least 50, so as long as $0.2 < \hat p < 0.8$, the conditions for the CLT (two sample; proportions) are satisfied. They may be satisfied for products with more extreme reorder proportions that have a greater number of product orders.

Choose the 5,000th, 10,000th, 15,000th, and 20,000th most reordered product (with at least 50 orders) to compare. In the code below we compute:

\begin{equation*}
\begin{array}{|l|c|c|c|c|}
\hline
& 5000 & \ 10000 & 15000 & 20000 \\ \hline
\text{mean} & \hat{p}_{5000} = 0.614286 & \hat{p}_{10000} = 0.530499 & \hat{p}_{15000} = 0.451327 & \hat{p}_{20000} = 0.353116 \\ \hline
\text{number} & n_{5000} = 70 & n_{10000} = 541 & n_{15000} = 113 & n_{20000} = 1011 \\ \hline
\text{reorders} & x_{5000} = 43 & x_{10000} = 287 & x_{15000} = 51 & x_{20000} = 357 \\ \hline
\end{array}
\end{equation*}


For $n \in \{5000, 10000, 15000, 20000\}$, let $p_n$ be the population reorder proportion. Form and test the hypotheses

$H_0: p_{n} = p_{m}$

$H_1: p_{n} \neq p_{m}$

where $n, m \in \{5000, 10000, 15000, 20000\}$, using frequentist tests at a level of significance $\alpha = 0.05$. Let's reveal what these four products are:

```python
alpha = 0.05
```

```python
mu_5000 = reorder_product_named['reordered_pct'].loc[5000]
num_5000 = reorder_product_named['total'].loc[5000]
re_5000 = round(mu_5000 * num_5000)
reorder_product_named.loc[5000]
```

The differences in reorder proportions for different products is often statistically and practically significant, justifying its use in making predictions:

* $(n, m) = (5000, 10000)$:

```python
print(reorder_product_named['product_name'].loc[5000] + ' was ordered %i times.\n\
Of those, %i were reorders. The proportion of reorders is %.4f.' % (num_5000, re_5000, mu_5000))

print(reorder_product_named['product_name'].loc[10000] + ' was ordered %i times.\n\
Of those, %i were reorders. The proportion of reorders is %.4f.' % (num_10000, re_10000, mu_10000))

print('The difference in proportions is %.4f.' % (mu_5000 - mu_10000))

pool = (re_5000 + re_10000) / (num_5000 + num_10000) # pooled proportion

z_stat = (mu_5000 - mu_10000) / math.sqrt(pool * (1-pool) * (1/num_5000 + 1/num_10000))
p_value = 2 * stats.norm.sf(abs(z_stat)) # Two-tail
print('The z-statistic is %.2f. and the p-value is %.5f.' % (z_stat, p_value))

if p_value > alpha:
    print('Fail to reject H_0.')
else:
    print('Reject H_0.')
```

##### Hypothesis Testing

Due to the large number of orders in our dataset, the differences in reorder proportions between different aisles is statistically significant, as long as the difference in reorder proportions is at least $0.02$, two percentage points. This follows from the identity we derived above,
\begin{equation*}
2 \sqrt{\hat{p} (1-\hat{p}) \left(n_1^{-1} + n_2^{-1}\right)} < \left| \hat{p}_1 - \hat{p}_2 \right|,
\end{equation*}
with $n_1 = n_2 = 5000$ (and $\hat{p} = 0.5$), which is a lower bound on the number of orders per aisle.


In order to construct a hypothesis test, compare the reorder proportions of two of the least frequented† aisles which, in addition, have a relatively small difference in reorder proportions. The `beauty` and `deodorants` aisles, for example, satisfy these conditions. These are also two samples which satisfy the conditions of the CLT for proportions. Smaller sample sizes and smaller differences between sample statistics yield a smaller (in magnitude) $z$-statistic; for such samples we are less likely to adopt $H_1$ (stated below) than for larger samples or samples with larger differences in reorder proportions. In the code below we compute:

\begin{equation*}
\begin{array}{|l|c|c|}
\hline
& \text{beauty} & \text{deodorants} \\ \hline
\text{mean} & \hat{p}_{\text{beauty}} = 0.2129 & \hat{p}_{\text{deodorants}} = 0.2388 \\ \hline
\text{number} & n_{\text{beauty}} = 6455 & n_{\text{deodorants}} = 15673 \\ \hline
\text{reorders} & x_{\text{beauty}} = 1374 & x_{\text{deodorants}} = 3743 \\ \hline
\end{array}
\end{equation*}


For the `beauty` population reorder proportion $p_{\text{beauty}}$ and the `deodorants` population reorder proportion $p_{\text{deodorants}}$, form and test the hypotheses

$H_0: p_{\text{beauty}} = p_{\text{deodorants}}$

$H_1: p_{\text{beauty}} \neq p_{\text{deodorants}}$

using frequentist tests at a level of significance $\alpha = 0.05$.




```python
alpha = 0.05
```

```python
d.groupby('aisle')['aisle_id'].count().sort_values().head(15)
```

Note that `beauty` is the second-least frequented aisle and `deodorants` are the 14th-least frequented aisle, out of 134 aisles.

```python
mu_beauty = d[d['aisle'] == 'beauty']['reordered'].mean()
num_beauty = len(d[d['aisle'] == 'beauty'])
re_beauty = len(d[(d['aisle'] == 'beauty') & (d['reordered'] == 1)])

print('The number of items ordered from the beauty aisle is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_beauty, re_beauty, mu_beauty))

mu_deodorants = d[d['aisle'] == 'deodorants']['reordered'].mean()
num_deodorants = len(d[d['aisle'] == 'deodorants'])
re_deodorants = len(d[(d['aisle'] == 'deodorants') & (d['reordered'] == 1)])

print('The number of items ordered from the deodorants aisle is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_deodorants, re_deodorants, mu_deodorants))

print('The difference in proportions is %.4f.' % (mu_beauty - mu_deodorants))

pool = (re_deodorants + re_beauty) / (num_deodorants + num_beauty) # pooled proportion

z_stat = (mu_deodorants - mu_beauty) / math.sqrt(pool * (1-pool) * (1/num_deodorants + 1/num_beauty))
p_value = 2 * stats.norm.sf(abs(z_stat)) # Two-tail
print('The z-statistic is %.2f. and the p-value is %.5f.' % (z_stat, p_value))

if p_value > alpha:
    print('Fail to reject H_0.')
else:
    print('Reject H_0.')
```

Even for two reorder proportions which are close in value $\left(\hat{p}_{\text{beauty}} - \hat{p}_{\text{deodorants}} = -0.026\right)$ and have small subsample sizes, the difference in proportions is significant. Given the reasoning above, this suggests that the difference in reorder proportions between most aisles is statistically significant. Therefore, we can justify using `aisle` to make predictions.

For reorder proportions which are very close in value, even for aisles that do not have a small number of orders, we can fail to reject $H_0$. To demonstrate, consider the `yogurt` and `cream` aisles.

$$
\begin{array}{|l|c|c|}
\hline
& \text{yogurt} & \text{cream} \\ \hline
\text{mean} & \hat{p}_{\text{yogurt}} = 0.6865 & \hat{p}_{\text{cream}} = 0.6852 \\ \hline
\text{number} & n_{\text{yogurt}} = 1507583 & n_{\text{cream}} = 330358 \\ \hline
\text{reorders} & x_{\text{yogurt}} = 1034957 & x_{\text{cream}} = 226356 \\ \hline
\end{array}
$$

Note that the difference in reorder proportions is $\hat{p}_{\text{yogurt}} - \hat{p}_{\text{cream}} = 0.0013.$

For the `yogurt` population reorder proportion $p_{\text{yogurt}}$ and the `cream` population reorder proportion $p_{\text{cream}}$, form and test the hypotheses

$H_0: p_{\text{yogurt}} = p_{\text{cream}}$

$H_1: p_{\text{yogurt}} \neq p_{\text{cream}}$

using frequentist tests at a level of significance $\alpha = 0.05$.

```python
mu_yogurt = d[d['aisle'] == 'yogurt']['reordered'].mean()
num_yogurt = len(d[d['aisle'] == 'yogurt'])
re_yogurt = len(d[(d['aisle'] == 'yogurt') & (d['reordered'] == 1)]) # == mu * num

print('The number of items ordered from the yogurt aisle is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_yogurt, re_yogurt, mu_yogurt))

mu_cream = d[d['aisle'] == 'cream']['reordered'].mean()
num_cream = len(d[d['aisle'] == 'cream'])
re_cream = len(d[(d['aisle'] == 'cream') & (d['reordered'] == 1)]) # == mu * num

print('The number of items ordered from the cream aisle is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_cream, re_cream, mu_cream))

print('The difference in proportions is %.4f.' % (mu_yogurt - mu_cream))

pool = (re_cream + re_yogurt) / (num_cream + num_yogurt) # pooled proportion

z_stat = (mu_cream - mu_yogurt) / math.sqrt(pool * (1-pool) * (1/num_cream + 1/num_yogurt))
p_value = 2 * stats.norm.sf(abs(z_stat)) # Two-tail
print('The z-statistic is %.2f. and the p-value is %.5f.' % (z_stat, p_value))

if p_value > alpha:
    print('Fail to reject H_0.')
else:
    print('Reject H_0.')
```

We cannot conclude that such a small difference in reorder proportions is statistically significant.

We can estimate from the above that, roughly speaking, as long as the difference in reorder proportions is $\gtrsim 0.01$, that difference is statistically significant. A $p$-value of $0.05$ corresponds to a difference in proportions of roughly $0.014$, assuming a proportion of $0.5$ and sample sizes of $10,000$.

##### Hypothesis Testing

Let's inspect the significance of differences in reorder ratio by time of day and day of week. If we presume the same conditions as in our above tests, then differences in reorder proportions between aisles are significant when the difference is at least $0.02$. This is because there are no `(order_dow, order_hour_of_day)` pairs for which the number of orders is less than 5000. We will use the same strategy that we used to examine the significance of differences in reorder ratio by aisle. Pick two pairs of `(order_dow, order_hour_of_day)` with a small difference in reorder proportion and a relatively small number of orders to potentially demonstrate significance for larger differences in reorder proportion and sample size.

```python
# Numerical values which are displayed in the above heatmap
reorder_hour_dow
```

```python
reorder_hour_dow.idxmin(axis=1)
```

```python
# Count of number of orders by (order_dow, order_hour_of_day)
reorder_hour_dow_count = d.groupby(['order_dow', 'order_hour_of_day'])['order_id'].aggregate('count').reset_index()
reorder_hour_dow_count = reorder_hour_dow_count.pivot('order_dow', 'order_hour_of_day', 'order_id')
reorder_hour_dow_count
```

```python
# Count of number of reorders by (order_dow, order_hour_of_day)
reorder_hour_dow_success = d.groupby(['order_dow', 'order_hour_of_day'])['reordered'].aggregate('sum').reset_index()
reorder_hour_dow_success = reorder_hour_dow_success.pivot('order_dow', 'order_hour_of_day', 'reordered')
reorder_hour_dow_success
```

* Tuesday midnight and Tuesday 1am, i.e. $(3,0)$ and $(3,1)$:

$$
\begin{array}{|l|c|c|}
\hline
& \text{(3,0)} & \text{(3,1)} \\ \hline

\text{number} & n_{\text{(3,0)}} = 28513 & n_{\text{(3,1)}} = 14205 \\ \hline
\text{reorders} & x_{\text{(3,0)}} = 15800 & x_{\text{(3,1)}} = 7607 \\ \hline
\end{array}
$$

```python
mu_31, mu_30 = reorder_hour_dow.loc[3,1], reorder_hour_dow.loc[3,0]
num_31, num_30 = reorder_hour_dow_count.loc[3,1], reorder_hour_dow_count.loc[3,0]
re_31, re_30 = reorder_hour_dow_success.loc[3,1], reorder_hour_dow_success.loc[3,0]

print('The number of items ordered during (3,1) is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_31, re_31, mu_31))

print('The number of items ordered during (3,0) is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_30, re_30, mu_30))

print('The difference in proportions is %.4f.' % (mu_30 - mu_31))

pool = ((mu_31 * num_31) + (mu_30 * num_30)) / (num_31 + num_30)

stderr = math.sqrt(pool * (1-pool) * (1/num_31 + 1/num_30))

print('The margin of error is %.5f' % (2 * stderr))

z_stat = (mu_30 - mu_31) / stderr

p_value = 2 * stats.norm.sf(abs(z_stat)) # Two-tail
print('The z-statistic is %.2f. and the p-value is %.5f.' % (z_stat, p_value))

if p_value > alpha:
    print('Fail to reject H_0.')
else:
    print('Reject H_0.')
```

Even for nearby means and small subsample sizes, the difference in reorder ratio at different times of day is significant, as long as that difference is at least roughly 0.01. Let's take a look at a difference in reorder ratios which is even smaller than that:

```python
mu_32, mu_31 = reorder_hour_dow.loc[3,2], reorder_hour_dow.loc[3,1]
num_32, num_31 = reorder_hour_dow_count.loc[3,2], reorder_hour_dow_count.loc[3,1]
re_32, re_31 = reorder_hour_dow_success.loc[3,2], reorder_hour_dow_success.loc[3,1]

print('The number of items ordered during (3,2) is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_32, re_32, mu_32))

print('The number of items ordered during (3,1) is %i.\n\
Of those, %i are reorders. The proportion of reorders is %.4f.' % (num_31, re_31, mu_31))

print('The difference in proportions is %.4f.' % (mu_31 - mu_32))

pool = ((mu_32 * num_32) + (mu_31 * num_31)) / (num_32 + num_31)

stderr = math.sqrt(pool * (1-pool) * (1/num_32 + 1/num_31))

print('The margin of error is %.5f' % (2 * stderr))

z_stat = (mu_31 - mu_32) / stderr

p_value = 2 * stats.norm.sf(abs(z_stat)) # Two-tail
print('The z-statistic is %.2f. and the p-value is %.5f.' % (z_stat, p_value))

if p_value > alpha:
    print('Fail to reject H_0.')
else:
    print('Reject H_0.')
```
